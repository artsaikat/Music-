```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit(df["description"])
```

```python
with open('count_vectorizer.pkl','wb') as f:
    pickle.dump(cv,f)
```

```python
def similar(text1,text2):
    v1 = cv.transform([text1])
    v2 = cv.transform([text2])

    score = cosine_similarity(v1, v2)

    return float(score[0][0])
```

```python
def recommend_songs(user_context,
                    final_emotion,
                    music_df):

    filtered = music_df[
        music_df["emotion"] == final_emotion
    ]

    scores = []

    for _, row in filtered.iterrows():

        score = similar(
            user_context,
            row["description"]
        )

        scores.append(
            (
                row["song"],
                score
            )
        )

    scores.sort(
        key=lambda x: x[1],
        reverse=True
    )

    return scores[:10]
```

```python
user_text = "i am feel very happy today"
```

```python
emotion, confidence = predict_emotion(user_text)
print(emotion, confidence)
```

```python
recommendations = recommend_songs(
    user_context=user_text,
    final_emotion=emotion,
    music_df=df
)
```

```python
print("\nRecommended Songs:\n")

for rank, (song, score) in enumerate(recommendations, start=1):
    print(f"{rank}. {song}")
```
